### Step 2.3 Perform the recursive feature elimination (RFE) technique to identify the effective features for building the model (10 marks)

```python
#For loop to get the accuracy score from 1 feature, then adding one feature in each step.
acc_scores = []
for i in range(1,57):
    clf = LogisticRegression()
    rfe = RFE(estimator=clf, n_features_to_select=i)
    # training model
    rfe.fit(X_train, y_train)
    # predicting on test set
    y_pred = rfe.predict(X_test)
    acc_score = accuracy_score(y_test, y_pred)
   
    print(f"Acc on test set using {i} features: {acc_score:.6f}")
    
    acc_scores.append(acc_score) # Store the scores in a list
```

```python
from sklearn.metrics import f1_score as sklearn_f1_score # renamed as i was having errors
#For loop to get the f1 score from 1 feature, then adding one feature in each step.
f1_scores = []
# For loop to get the f1 score by selecting 1 to 56 features
for i in range(1, 57):
    clf = LogisticRegression()
    rfe = RFE(estimator=clf, n_features_to_select=i)
    rfe.fit(X_train, y_train)
    y_pred = rfe.predict(X_test)
    
    # Calculate the F1 score
    f1 = sklearn_f1_score(y_test, y_pred)
    # Store the F1 score
    f1_scores.append(f1)
    
    # Print the F1 score for the current number of features
    print(f"F1 score on test set using {i} features: {f1:.6f}")
```
