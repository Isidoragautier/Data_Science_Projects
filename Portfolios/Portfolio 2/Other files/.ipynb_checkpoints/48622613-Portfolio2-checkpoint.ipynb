### Q3: Split Training and Testing Data
* Machine learning models are trained to help make predictions for the future. Normally, we need to randomly split the dataset into training and testing sets, where we use the training set to train the model, and then leverage the well-trained model to make predictions on the testing set.
* To further investigate whether the size of the training/testing data affects the model performance, please randomly split the data into training and testing sets with different sizes:
    * Case 1: training data containing 10% of the entire data;
    * Case 2: training data containing 90% of the entire data.
* Print the shape of training and testing sets in the two cases.

```python
train1, test1 = train_test_split(car, test_size=0.9, random_state=142) # 10% training data, 90% test data
train2, test2 = train_test_split(car, test_size=0.1, random_state=142) # 90% training data, 10% test data

print("Case 1 = Train:", train1.shape, "Test:", test1.shape)
print("Case 2 = Train:", train2.shape, "Test:", test2.shape)
```

### Q4: Train Linear Regression Models with Feature Selection under Cases 1 & 2
* When training a machine learning model for prediction, we may need to select the most important/correlated input features for more accurate results.
* To investigate whether feature selection affects the model performance, please select two most correlated features and two least correlated features regarding selling_price, respectively.
* Train four linear regression models by following the conditions:
    - (model-a) using the training/testing data in case 1 with two most correlated input features
    - (model-b) using the training/testing data in case 1 with two least correlated input features
    - (model-c) using the training/testing data in case 2 with two most correlated input features
    - (model-d) using the training/testing data in case 2 with two least correlated input features
* By doing this, we can verify the impacts of the size of traing/testing data on the model performance via comparing model-a and model-c (or model-b and model-d); meanwhile the impacts of feature selection can be validated via comparing model-a and model-b (or model-c and model-d).    

#### **model-a**:  using the training/testing data in case 1 with two most correlated input features
Linear regression model selling price with Transmission code and year features:

```python
train1_x = train1[['year','fuel_code']]
train1_y = train1['selling_price']
test1_x= test1[['year','fuel_code']]
test1_y = test1['selling_price']
reg1 = linear_model.LinearRegression()
reg1.fit(train1_x, train1_y)
print("y = x *", reg1.coef_, "+", reg1.intercept_)
```

#### **model-b**:  using the training/testing data in case 1 with two least correlated input features
Linear regression model selling price with km_driven and owner code features:

```python
train2_x = train1[['km_driven', 'owner_code']]
train2_y = train1['selling_price']
test2_x= test1[['km_driven', 'owner_code']]
test2_y = test1['selling_price']
reg2 = linear_model.LinearRegression()
reg2.fit(train2_x, train2_y)
print("y = x *", reg2.coef_, "+", reg2.intercept_)
```

#### **model-c** using the training/testing data in case 2 with two most correlated input features
Linear regression model selling price with Transmission code and year features:

```python
train3_x = train2[['year','fuel_code']]
train3_y = train2['selling_price']
test3_x= test2[['year','fuel_code']]
test3_y = test2['selling_price']
reg3 = linear_model.LinearRegression()
reg3.fit(train1_x, train1_y)
print("y = x *", reg3.coef_, "+", reg3.intercept_)
```

#### **model-d** using the training/testing data in case 2 with two least correlated input features
Linear regression model selling price with km_driven and owner code features:

```python
train4_x = train2[['km_driven', 'owner_code']]
train4_y = train2['selling_price']
test4_x= test2[['km_driven', 'owner_code']]
test4_y = test2['selling_price']
reg4 = linear_model.LinearRegression()
reg4.fit(train4_x, train4_y)
print("y = x *", reg4.coef_, "+", reg4.intercept_)
```

### Q5: Evaluate Models
* Evaluate the performance of the four models with two metrics, including MSE and Root MSE
* Print the results of the four models regarding the two metrics

```python
predicted1 = reg1.predict(test1_x)
mse1 = ((np.array(test1_y) - predicted1)**2).sum() / len(test1_y)
r2_1 = r2_score(test1_y, predicted1)
rmse1 = np.sqrt(mse1)
print("Model a: 10% Training Data / Selling price vs Fuel code and Year")
print(f"MSE model a: {mse1:,.2f}")
print(f"Root MSE model a: {rmse1:,.2f}")
print(f"R Squared model a: {r2_1:.3f}")
print("model a: y = x *", reg1.coef_, "+", reg1.intercept_)
print("==================================================")

predicted2 = reg2.predict(test2_x)
mse2 = ((np.array(test2_y) - predicted2)**2).sum() / len(test2_y)
r2_2 = r2_score(test2_y, predicted2)
rmse2 = np.sqrt(mse2)
print("Model b: 10% Training Data / Selling price vs Km_driven and Owner code")
print(f"MSE model b: {mse2:,.2f}")
print(f"Root MSE model b: {rmse2:,.2f}")
print(f"R Squared model b: {r2_2:.3f}")
print("model b: y = x *", reg2.coef_, "+", reg2.intercept_)
print("==================================================")

predicted3 = reg3.predict(test3_x)
mse3 = ((np.array(test3_y) - predicted3)**2).sum() / len(test3_y)
r2_3 = r2_score(test3_y, predicted3)
rmse3 = np.sqrt(mse3)
print("Model c: 90% Training Data / Selling price vs Fuel code and Year")
print(f"MSE model c: {mse3:,.2f}")
print(f"Root MSE model c: {rmse3:,.2f}")
print(f"R Squared model c: {r2_3:.3f}")
print("model c: y = x *", reg3.coef_, "+", reg3.intercept_)
print("==================================================")

predicted4 = reg4.predict(test4_x)
mse4 = ((np.array(test4_y) - predicted4)**2).sum() / len(test4_y)
r2_4 = r2_score(test4_y, predicted4)
rmse4 = np.sqrt(mse4)
print("Model d: 90% Training Data / Selling price vs Km_driven and Owner code")
print(f"MSE model d: {mse4:,.2f}")
print(f"Root MSE model d: {rmse4:,.2f}")
print(f"R Squared model d: {r2_4:.3f}")
print("model d: y = x *", reg4.coef_, "+", reg4.intercept_)
```

### Q6: Visualize, Compare and Analyze the Results
* Visulize the results, and perform ___insightful analysis___ on the obtained results. For better visualization, you may need to carefully set the scale for the y-axis.
* Normally, the model trained with most correlated features and more training data will get better results. Do you obtain the similar observations? If not, please ___explain the possible reasons___.

```python
# Create a data frame with the results 
regression_results = {
    'Model': ['Model a', 'Model b', 'Model c', 'Model d'],
    'MSE': [mse1, mse2, mse3, mse4],
    'RMSE': [rmse1, rmse2, rmse3, rmse4],
    'R^2': [r2_1, r2_2, r2_3, r2_4]
}
df_regression_results = pd.DataFrame(regression_results)

# Sort the bars to easier understanting.
df_sorted_mse = df_regression_results.sort_values(by='MSE', ascending=False)
df_sorted_rmse = df_regression_results.sort_values(by='RMSE', ascending=False)
df_sorted_r2 = df_regression_results.sort_values(by='R^2', ascending=True)

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

#Plot each result in a different plot:
axes[0].bar(df_sorted_mse['Model'], df_sorted_mse['MSE'], color='blue')
axes[0].set_title('Mean Squared Error (MSE)')
axes[0].set_ylabel('MSE')
axes[0].set_ylim(bottom=1.8e11)

axes[1].bar(df_sorted_rmse['Model'], df_sorted_rmse['RMSE'], color='green')
axes[1].set_title('Root Mean Squared Error (RMSE)')
axes[1].set_ylabel('RMSE')
axes[1].set_ylim(bottom=420000)

axes[2].bar(df_sorted_r2['Model'], df_sorted_r2['R^2'], color='red')
axes[2].set_title('R Squared (R^2)')
axes[2].set_ylabel('R^2')

plt.tight_layout()
plt.show()
```

**Findings:**

1. **Model Performance**: Based on the MSE and RMSE metrics, Model C demonstrates the best performance with the lowest error values, indicating the best fit to the target variable. Model A follows in performance, with Model D next, and Model B showing the highest error. These results indicate that Model C is the most accurate, while Model B is the least accurate.

2. **Impact of Features and Training Data**: Model C is superior due to its use of the two most correlated features with the selling price and being trained on 90% of the data. This highlights the importance of both feature relevance and a substantial training dataset for achieving better model performance.

3. **Data Size Considerations**: Although Model A incorporates the most relevant features, it is trained on only 10% of the data (365 rows). This limited data size likely hampers its performance compared to Model C, which utilizes 90% of the data (3291 rows), impacting Model A's ability to generalize effectively.

4. **Low R^2 Values**: All models exhibit relatively low R^2 values, with the highest being 0.23. This suggests that the current set of features does not fully explain the variability in the selling price. There may be other relevant features that could enhance predictive accuracy.

5. **Model B and D Performance**: Models B and D show the highest errors and the lowest R^2 values, indicating poorer performance compared to Models A and C. This is likely due to the variables used in these models being less correlated with the selling price.

**Conclusion:**

The analysis indicates that Model C is the most effective model among those built, benefiting from both highly correlated features and a larger training dataset. Model A, while also using relevant features, suffers from limited data. Models B and D, with their higher errors and lower R² values, are less effective due to their inclusion of less relevant features. The generally low R² values across all models suggest that there may be additional features that could improve the prediction of selling prices.
